```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/nlp cODE/Data_nlp.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

```python
data=train.iloc[:,2:27]
data.replace(['^a-zA-Z'],' ',regex=True,inplace=True)

list1=[i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns=new_Index
```

```python
data.head()
```

```python
for index in new_Index:
  data[index]=data[index].str.lower()
  
```

```python
data.head()
```

```python
headlines=[]
for row in range(0,len(data.index)):
  headlines.append(" ".join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
count_vector=CountVectorizer(ngram_range=(2,2))
```

```python
train_ds=count_vector.fit_transform(headlines)
```

```python
Randomfrst=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
Randomfrst.fit(train_ds,train['Label'])
```

```python
test_transfrom=[]
for row in range(0,len(test.index)):
  test_transfrom.append(' '.join(str(x) for  x in test.iloc[row,2:27]))
test_transfrom=count_vector.transform(test_transfrom)
prediction=Randomfrst.predict(test_transfrom)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],prediction)
print(matrix)
```

```python
accuracy_score(test['Label'],prediction)
```

```python
report=classification_report(test['Label'],prediction)
```

```python
print(report)
```
